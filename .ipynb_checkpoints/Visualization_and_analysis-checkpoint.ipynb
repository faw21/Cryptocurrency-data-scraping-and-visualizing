```python
import numpy as np
import pandas as pd
from statsmodels.tsa.stattools import adfuller
import matplotlib
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
#import data
#data of 2 years
ETH2Y = pd.read_csv('ETH-USD-2year.csv', parse_dates = ['Date'])
BTC2Y = pd.read_csv('BTC-USD-2year.csv', parse_dates = ['Date'])
#data of 6 months
ETH6M = pd.read_csv('ETH-USD-6month.csv', parse_dates = ['Date'])
BTC6M = pd.read_csv('BTC-USD-6month.csv', parse_dates = ['Date']) 
```

```python
#delete the adjusting close price from each data frame
del ETH2Y['Adj Close']
del BTC2Y['Adj Close']
del ETH6M['Adj Close']
del BTC6M['Adj Close']
```

```python
#set datatime as index
ETH2Y.set_index('Date', inplace=True)
BTC2Y.set_index('Date', inplace=True)
ETH6M.set_index('Date', inplace=True)
BTC6M.set_index('Date', inplace=True)
```

```python
#update dataframes, add a new column of daily return
ETH2Y = ETH2Y.assign(DailyReturnPct = lambda x: ((x['Close']-x['Open'])/x['Open']*100))
BTC2Y = BTC2Y.assign(DailyReturnPct = lambda x: ((x['Close']-x['Open'])/x['Open']*100))
ETH6M = ETH6M.assign(DailyReturnPct = lambda x: ((x['Close']-x['Open'])/x['Open']*100))
BTC6M = BTC6M.assign(DailyReturnPct = lambda x: ((x['Close']-x['Open'])/x['Open']*100))
```

```python
#visulize data
#the close price is now in column No.3
ETH2Y.iloc[:,3].plot()
plt.title('Last 2 years of ETH Price')
plt.show()

BTC2Y.iloc[:,3].plot()
plt.title('Last 2 years of BTC Price')
plt.show()

ETH6M.iloc[:,3].plot()
plt.title('Last 6 months of ETH Price')
plt.show()

BTC6M.iloc[:,3].plot()
plt.title('Last 6 months of BTC Price')
plt.show()
```

```python
#print information
#average price and daily return of each crypto for last 2 years
#std of price and daily return of each crypto for last 2 years
#correlation of them
print('For the recent 2 years: \n')
print('The average closing price of ETH is: ' + str(ETH2Y.iloc[:,3].mean()))
print('The std of closing price of ETH is: ' + str(ETH2Y.iloc[:,3].std()))
print()
print('The average closing price of BTC is: ' + str(BTC2Y.iloc[:,3].mean()))
print('The std of closing price of BTC is: ' + str(BTC2Y.iloc[:,3].std()))
print()
print('The correlation between BTC\'s and ETH\'s closing price is: ' + str(BTC2Y.iloc[:,3].corr(ETH2Y.iloc[:,3])))

print()
print('The average daily return(%) of ETH is: ' + str(ETH2Y.iloc[:,5].mean()))
print('The std of daily return(%) of ETH is: ' + str(ETH2Y.iloc[:,5].std()))
print()
print('The average daily return(%) of BTC is: ' + str(BTC2Y.iloc[:,5].mean()))
print('The std of daily return(%) of BTC is: ' + str(BTC2Y.iloc[:,5].std()))
print()
print('The correlation between BTC\'s and ETH\'s daily return(%) is: ' + str(BTC2Y.iloc[:,5].corr(ETH2Y.iloc[:,5])))
```

```python
#average price and daily return of each crypto for last 6 months
#std of price and daily return of each crypto for last 6 months
#correlation of them
print('For the recent 6 months: \n')
print('The average closing price of ETH is: ' + str(ETH6M.iloc[:,3].mean()))
print('The std of closing price of ETH is: ' + str(ETH6M.iloc[:,3].std()))
print()
print('The average closing price of BTC is: ' + str(BTC6M.iloc[:,3].mean()))
print('The std of closing price of BTC is: ' + str(BTC6M.iloc[:,3].std()))
print()
print('The correlation between BTC\'s and ETH\'s closing price is: ' + str(BTC6M.iloc[:,3].corr(ETH6M.iloc[:,3])))

print()
print('The average daily return(%) of ETH is: ' + str(ETH6M.iloc[:,5].mean()))
print('The std of daily return(%) of ETH is: ' + str(ETH6M.iloc[:,5].std()))
print()
print('The average daily return(%) of BTC is: ' + str(BTC6M.iloc[:,5].mean()))
print('The std of daily return(%) of BTC is: ' + str(BTC6M.iloc[:,5].std()))
print()
print('The correlation between BTC\'s and ETH\'s daily return(%) is: ' + str(BTC6M.iloc[:,5].corr(ETH6M.iloc[:,5])))
```

```python
#Dickey Fuller test for each crypto for recent 2 years
print('For the recent 2 years:')

adf_test = adfuller(BTC2Y.iloc[:,3], maxlag = 0)
p_value = adf_test[1]

print('\tFrom the Dickey-Fuller test, the series of closing price of Bitcoin has a p-value of ' + str(p_value))
if p_value > 0.05:
    print('\tSince p-value is greater than 0.05, the price is not stationary.')
else:
    print('\tSince p-value is less than or equal to 0.05, the price is stationary.')
    
adf_test = adfuller(ETH2Y.iloc[:,3], maxlag = 0)
p_value = adf_test[1]

print('\tFrom the Dickey-Fuller test, the series of closing price of Ethereum has a p-value of ' + str(p_value))
if p_value > 0.05:
    print('\tSince p-value is greater than 0.05, the price is not stationary.')
else:
    print('\tSince p-value is less than or equal to 0.05, the price is stationary.')
```

```python
#Dickey Fuller test for each crypto for recent 6 months
print('For the recent 6 months:')

adf_test = adfuller(BTC6M.iloc[:,3], maxlag = 0)
p_value = adf_test[1]

print('\tFrom the Dickey-Fuller test, the series of closing price of Bitcoin has a p-value of ' + str(p_value))
if p_value > 0.05:
    print('\tSince p-value is greater than 0.05, the price is not stationary.')
else:
    print('\tSince p-value is less than or equal to 0.05, the price is stationary.')
    
adf_test = adfuller(ETH6M.iloc[:,3], maxlag = 0)
p_value = adf_test[1]

print('\tFrom the Dickey-Fuller test, the series of closing price of Ethereum has a p-value of ' + str(p_value))
if p_value > 0.05:
    print('\tSince p-value is greater than 0.05, the price is not stationary.')
else:
    print('\tSince p-value is less than or equal to 0.05, the price is stationary.')
```
